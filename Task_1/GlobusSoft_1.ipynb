# Scraping Data

```python
#Defining the empty array to store scrap data
name = []
storage = []
color = []
img = []
price = []
```

```python
#Getting the counter for number of data points present
tags = driver.find_elements_by_xpath('//div[@class="mnr-c pla-unit"]/a[2]')
len(tags)
```

```python
#Scraping image link
tag01 = driver.find_elements_by_xpath('//div[@class="Ter3Ue Nwl2ud"]')
for i1 in tag01:
    j1 = i1.find_element_by_xpath('//a[2]/div[1]/div[1]/img[1]')
    img.append(j1.get_attribute('src'))
```

```python
len(img)
```

```python
img
```

```python
#Scraping model name
tag02 = driver.find_elements_by_xpath('//span[@class="pymv4e"]')
for i2 in tag02:
    if(i2.text!=""):
        name.append(i2.text.split("(")[0].split("|")[0].split(",")[0])
```

```python
len(name)
```

```python
name
```

```python
#Scraping price
tag03 = driver.find_elements_by_xpath('//span[@class="e10twf"]')
for i in tag03:
    price.append(i.text.split("₹")[1])
```

```python
len(price)
```

```python
price
```

```python
#Scraping the web links
link=[]
for i in tags:
    link.append(i.get_attribute('href'))
```

```python
len(link)
```

```python
link
```

```python
#Scraping the storage value
for url in link:
    driver.get(url)
    
    try:
        tag2 = driver.find_element_by_xpath('//div[@id="option-device-memory-IM003"]/div[1]/ul[1]/li[1]/div[1]/label[1]/span[1]/span[1]/strong[1]')
        storage.append(tag2.text.split("|")[0])
    except:
        try:
            tag2 = driver.find_element_by_xpath('//div[@class="cp-specification"]/ul[4]/li[2]/ul[1]/li[2]')
            storage.append(tag2.text.split("|")[0])
        except:
            tag2 = driver.find_element_by_xpath('//div[@id="storage"]/div[2]/ul[1]/li[1]/div[1]/label[1]/span[2]/span[1]/span[1]')
            storage.append(tag2.text.split("|")[0])
```

```python
len(storage)
```

```python
storage
```

```python
#Scraping for color option
for url in link:
    driver.get(url)
    
    try:
        tag4 = driver.find_element_by_xpath('//span[@id="multiColorText"]')
        color.append(tag4.text)
    except:
        try:
            tag4 = driver.find_element_by_xpath('//div[@class="cp-specification"]/ul[18]/li[2]/ul[1]/li[2]')
            color.append(tag4.text)
        except:
            try:
                tag4 = driver.find_element_by_xpath('//label[@for="OPT-001_8"]/span[2]')
                color.append(tag4.text)
            except:
                tag4 = driver.find_element_by_xpath('//div[@class="hubble-product__options-color"]/ul[1]/li[1]/div[1]/label[1]/span[2]')
                color.append(tag4.text)
```

```python
len(color)
```

```python
#Check on data if complete to create a dataframe
print(len(name),len(storage),len(color),len(img),len(price))
```

# Data Frame

```python
#Creating a DataFrame of the scraped data
Data_Ad_Scrap = pd.DataFrame({"Model"       :  name,
                              "Storage"     :  storage,
                              "Color"       :  color,
                              "Image Link"  :  img,
                              "Web Link"    :  link,
                              "Price"       :  price
                             })
```

```python
#Calling the dataframe created to view
Data_Ad_Scrap
```

# Timestamp Creation

```python
#Producing file name with timestamp
t = time.localtime()
timestamp = time.strftime('%d-%b-%Y_%H%M', t)
file = ("Data-" + timestamp + ".csv")
```

# Saving the Dataframe

```python
#Saving to a .csv file
Data_Ad_Scrap.to_csv(file)
```
